```python
import matplotlib.pyplot as plt
import csv
```

```python
list_category = ['birds','clouds','music','planet','snake']

birds = []
clouds = []
music = []
planet = []
snake = []

list_of_normalized_lists = []

for this_category in list_category:
    with open('category-wise/'+ this_category + '/outputs_inter_over_Gsk.csv', 'r+') as csvfile:
        # Create a CSV reader object
        reader = csv.reader(csvfile)
        
        # csvfile.seek(0)
        # Find the maximum value in the column
        temp = []
        for row in reader:
            temp.append(float(row[0]))  # assuming the values are in the first column

        # print(temp)
        temp.reverse()
        list_of_normalized_lists.append(temp)



# print(final_normalized)


final_normalized = []

# Traverse the columns
for i in range(len(list_of_normalized_lists[0])):
  
  # Initialize a variable to store the column sum
  column_sum = 0
  
  # Traverse the rows and add up the column values
  for j in range(len(list_of_normalized_lists)):
    column_sum += list_of_normalized_lists[j][i]
  
  # Calculate the average and append it to the column_averages list
  column_average = column_sum / len(list_of_normalized_lists)
  final_normalized.append(column_average)

# Print the column averages
print(final_normalized)

# plt.draw(final_normalized)
plt.plot(final_normalized)
```

```python
# for this_category in list_category:
with open('category-wise/birds/outputs_inter_over_Gsk.csv', 'r+') as csvfile:
    # Create a CSV reader object
    reader = csv.reader(csvfile)
    
    # csvfile.seek(0)
    # Find the maximum value in the column
    column_values = []
    for row in reader:
        column_values.append(float(row[0]))  # assuming the values are in the first column

    max_value = max(column_values)
    
    print(column_values)
    # Reset the file pointer to the beginning of the file
    csvfile.seek(0)
    
    # Create a CSV writer object
    updated_rows = []
    
    # Divide each value in the column by the maximum value
    for row in reader:
        value = float(row[0])  # assuming the values are in the first column
        new_value = value / max_value
        updated_row = [new_value] + row[1:]  # assuming there are additional columns after the value column
        updated_rows.append(updated_row)
    
# Open the CSV file again for writing
with open('category-wise/birds/outputs_inter_over_Gsk.csv', 'w', newline='') as csvfile:
    # Create a CSV writer object
    writer = csv.writer(csvfile)
    
    # Write the updated rows to the file
    for row in updated_rows:
        writer.writerow(row)  # assuming there are additional columns after the value column
    
    # Truncate the file to the current position (to remove any extra rows)
    # csvfile.truncate()
```
